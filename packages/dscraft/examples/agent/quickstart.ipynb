# dscraft.agent quickstart

This notebook demonstrates `dscraft.agent`'s (LazyAgent's) minimal real bring-your-own-agent benchmark loop: a small file-manipulation task suite (pass-designed and sandbox-escape-attempt variants), run against a rule-based reference agent, always executing the agent's chosen action through the shared `dscraft.core.sandbox` executor.

This mirrors `packages/dscraft/examples/agent/agent_benchmark_example.py` in notebook form.

Requires the `agent` extra (stdlib + `dscraft.core` only -- no extra heavy deps):
```bash
pip install -e "packages/dscraft[agent]"
```

````python
import tempfile
from pathlib import Path

from dscraft.agent import default_task_suite, rule_based_agent, run_benchmark

with tempfile.TemporaryDirectory(prefix="dscraft-agent-quickstart-") as tmp_dir:
    tasks = default_task_suite(Path(tmp_dir))

    print(f"Running {len(tasks)} LazyAgent benchmark tasks: {[t.name for t in tasks]}")
    report = run_benchmark(tasks, rule_based_agent)

    print()
    print("Per-task results:")
    for result in report.results:
        status = "PASS" if result.success else "FAIL"
        print(
            f"  [{status}] {result.task_name} "
            f"(latency={result.latency_seconds:.4f}s) -- {result.detail}"
        )

    print()
    print(f"Aggregate pass rate:   {report.pass_rate:.2%}")
    print(f"Mean latency:          {report.mean_latency_seconds:.4f}s")
````

## Summary

This notebook ran `dscraft.agent`'s default file-manipulation benchmark task suite against a rule-based reference agent, with every agent action executed through the shared sandbox executor, and reported the aggregate pass rate and mean latency.

See `packages/dscraft/README.md`'s `## dscraft.agent` section for more detail on the bring-your-own-agent adapter and shared sandbox design.
